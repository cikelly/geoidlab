```python
import sys
sys.path.append('../src')
from legendre import legendre_poly
import kernels
# from utils.coordinates import geo_lat2geocentric
import numpy as np

import matplotlib.pyplot as plt
plt.rcParams.update({'font.size':11, 'font.family':'Arial'})
```

```python
t     = np.linspace(-1,1,500)
N     = 10
Pn    = np.zeros((len(t), N))


Pn[:,0] = 1
Pn[:,1] = t

for n in range(2, N):
    Pn[:,n] = ( (2*n-1)*t*Pn[:,n-1] - (n-1)*Pn[:,n-2] ) / n
```

```python
# Plot all lines at once
plt.plot(t, Pn[:,0:6], linewidth=0.65)

# Define the labels

labels = [f'P$_{str(i)}(x)$' for i in range(6)]

plt.grid(which='both', linewidth=0.2, color='gray')
plt.minorticks_on()
plt.grid(which='minor', linewidth=0.08, color='gray')

plt.ylim([-1, 1])
plt.xlim([-1, 1])


plt.legend(labels, fontsize=9, ncols=2)

plt.title('Legendre polynomials', fontweight='bold')
plt.show()
```
